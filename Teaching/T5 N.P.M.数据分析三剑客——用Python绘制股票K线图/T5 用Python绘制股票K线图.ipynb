### 股票的连续复利收益率

#### 理财产品的连续复利

**每复利一次可以认为是进行了再投资**

年利率3.5%的理财产品投资两年  

不复利：$R = 1 + 3.5\% \times 2$    
按每年复利：$R = (1 + 3.5\%)^2$    
按半年复利：$R = (1 + \frac{3.5\%}{2})^4 $  
按每月复利：$R = (1 + \frac{3.5\%}{12})^{24} $  
按每日复利：$R = (1 + \frac{3.5\%}{365})^{730} $  
按须臾复利：$R = \lim\limits_{x\rightarrow\infty}(1 + \frac{3.5\%}{x})^{2x} = e^{2\times3.5\%}$

```python
print('不复利：',1+0.035*2)
print('按每年复利：',(1+0.035)**2)
print('按半年复利：',(1+0.035/2)**4)
print('按每月复利：',(1+0.035/12)**24)
print('按每日复利：',(1+0.035/365)**730)
print('按须臾复利：',np.exp(2*0.035))
```

compouded continuously 连续复利  
compouded quarterly  季度复利

#### 股票复利的另一些原因

石川：金融计算收益率的时候为什么大部分用对数收益率 (Log Return) 而不是用算数收益率？ https://www.zhihu.com/question/30113132/answer/431786548

##### 为了反应真实收益率

今年涨了50%，明天跌了50%，最后盈利多少？

算术平均值： 0

```python
.5 + (-.5) 
```

但真实的情况： - 25%

```python
(1 + .5) * (1 - .5) - 1
```

对数收益率由于具备可加性，它的均值可以正确反映出该投资品的真实收益率

$R_t = ln(\frac{P_t}{P_{t-1}})$

$R_t = ln(\frac{P_t}{P_{t-1}}), R_{t-1} = ln(\frac{P_{t-1}}{P_{t-2}}) \dots,  R_{1} = ln(\frac{P_{1}}{P_{0}})$

$R_T = R_t + R_{t-1} + \dots + R_1 = ln(\frac{P_{t}}{P_{t-1}} \times \frac{P_{t-1}}{P_{t-2}} \times \dots \times \frac{P_{1}}{P_{0}}) = ln(\frac{P_{t}}{P_{0}})$

**使用对数的形式计算股票的真实收益率肯定会比直接算收益率要小**

```python
x = np.linspace(0, 10, 100)[1:]
y1 = np.log(x)
y2 = x - 1
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

##### 可以进行期望估计

如果能假设不同期是相互独立的，T 期对数收益率相加相当于 T 个独立的随机变量相加。由中心极限定理可知，它们的和逼近正态分布。

由大数定律可知，$(1/T) × ln(X_T/X_0)$ ，即单期对数收益率均值，随着 T 的增大一定会收敛于它的期望；对于给定的 T，T 期的总收益会收敛于 $E[ln(X_T/X_0)]$。


必如此处我们的T就是2022年1月到5月，由于T比较大，我们可以认为贵州茅台的股价这个对象的每日的对数收益率就是$(1/T) × ln(X_T/X_0)$，可以用于建模的期望估计

## K线图制作

k线之所以叫k线，是因为K线图这种图表，出自日本，在日本的“K”写做“罫”（日本音读kei），K线图称为“罫线”，后来西方以其英文首字母“K”直译为“K”线，由此发展而来。

K线图这种图表源处于日本德川幕府时代，被当时日本米市的商人用来记录米市的行情与价格波动，后因其细腻独到的标画方式而被引入到股市及期货市场。目前，这种图表分析法在我国以至整个东南亚地区均尤为流行。

如果使用 matplotlib 从 0 开始绘制，一步一步添加日线、均线、MACD、成交量等指标时，则会显得十分麻烦，且代码很难复用。  
matplotlib 中提供接口（matplotlib.finance）直接绘制K线，现在 matplotlib.finance 已经独立成库 mplfinance，更方便的让我们使用。

#### 做出符合国内习惯的蜡烛图

CSDN-Shepherdppz：全功能动态交互式K线图(蜡烛图)：https://blog.csdn.net/Shepherdppz/article/details/117575286

```python
import mplfinance as mpf
# 设置mplfinance的蜡烛颜色，up为阳线颜色，down为阴线颜色
my_color = mpf.make_marketcolors(up='r',
                                 down='g',
                                 edge='inherit',
                                 wick='inherit',
                                 volume='inherit')
# 设置图表的背景色
my_style = mpf.make_mpf_style(marketcolors=my_color,
                              figcolor='(0.82, 0.83, 0.85)',
                              gridcolor='(0.82, 0.83, 0.85)')
```

```python
mpf.plot(gzmt, style=my_style, type='candle', volume=True)
```

#### pyecharts实现

```python
from pyecharts import options as opts
from pyecharts.charts import Kline
```

```python
gzmt_echarts = gzmt.loc[:, ['open', 'close', 'low', 'high']]
```

```python
k = Kline()
```

```python
from pyecharts.globals import CurrentConfig, NotebookType 
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

k.add_xaxis(gzmt.index.astype('str').tolist())
k.add_yaxis("贵州茅台", gzmt_echarts.values.tolist())

k.set_global_opts(
    yaxis_opts=opts.AxisOpts(
        is_scale=True,
        splitarea_opts=opts.SplitAreaOpts(
            is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
        ),
    ),
    xaxis_opts=opts.AxisOpts(is_scale=True),
    datazoom_opts=[opts.DataZoomOpts(type_="inside")],
    title_opts=opts.TitleOpts(title="贵州茅台的蜡烛图"),

)

k.load_javascript()
```

```python
k.render_notebook()
```
