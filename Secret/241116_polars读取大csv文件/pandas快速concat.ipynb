## 创建数组

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col, row = int(3e2), int(4e5)
size = col * row
times = 10
```

```python
df = pd.DataFrame(np.arange(size).reshape(row, col))
```

```python
df
```

## 低效合并

```python
%%time
total_df = pd.DataFrame()
for _ in range(times):
    total_df = pd.concat([total_df, df])
```

```python
total_df
```

## 高效合并

```python
%%time
total_df_list = []
for _ in range(times):
    total_df_list.append(df)
pd.concat(total_df_list, axis=0)
```
